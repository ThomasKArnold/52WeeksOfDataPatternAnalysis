## Plot the datasets

### Get the Male Age Crime Curve Data

```python
MaleAgeCrimeCurvedf = pd.read_excel('MaleAgeCrimeCurve.xlsx')
MaleAgeCrimeCurvedf
```

### Plot the Male Age Crime Rate by Year

```python
# Plot creation
s = sns.set_context("paper")

plt.rc('axes', labelsize=10)    # fontsize of the x and y labels
plt.rc('axes', titlesize=10)  # fontsize of the figure title
plt.rc('xtick', labelsize=8)    # fontsize of the tick labels
plt.rc('ytick', labelsize=8)    # fontsize of the tick labels
plt.rc('legend', fontsize=10)    # legend fontsize

fig, ax = plt.subplots(figsize=(3.75,2), dpi=300, facecolor='w')

lin1 = ax.plot('Age', 'Crime Rate', data=MaleAgeCrimeCurvedf,color="blue", linewidth=2 )

ax.set_xlabel('Age')
ax.set_ylabel('Crime Rate')

ax.margins(x = 0.2, y= .01)
# plt.subplots_adjust(left=0, bottom=0, right=2, top=1, wspace=0, hspace=0)

#ax.tick_params(axis='y', colors='red')
plt.xlim([0, 101])
plt.ylim([0, .065])
plt.title('Male Age Crime Curve')

plt.savefig('MaleAgeCrimeCurve.png', dpi=300)

plt.show()
```

### Get the Medical Percentile Data

```python
MedicalCostsdf = pd.read_excel('MedicalPaymentPercentiles.xlsx')
MedicalCostsdf
```

### Plot the Average Annual Cost by Percentile

```python
# Plot creation
s = sns.set_context("paper")

plt.rc('axes', labelsize=9)    # fontsize of the x and y labels
plt.rc('axes', titlesize=9)  # fontsize of the figure title
plt.rc('xtick', labelsize=7)    # fontsize of the tick labels
plt.rc('ytick', labelsize=7)    # fontsize of the tick labels
plt.rc('legend', fontsize=10)    # legend fontsize

fig, ax = plt.subplots(figsize=(3.75,2), dpi=300, facecolor='w')

lin1 = ax.plot('PaymentPercentileInteger', 'PaymentAmount', data=MedicalCostsdf,color="blue", linewidth=3 )

ax.set_xlabel('Payment Percentile')
ax.set_ylabel('Average Annual Cost',color="blue")

ax.margins(x = 0.2, y= .01)
# plt.subplots_adjust(left=0, bottom=0, right=2, top=1, wspace=0, hspace=0)

ax.tick_params(axis='y', colors='blue')
plt.xlim([0, 102])
plt.ylim([0, 600001])
plt.title('Average Annual Medical Costs by Percentile', fontsize=9)

plt.savefig('AverageAnnualMedicalCostsbyPercentile.png', dpi=300)

plt.show()
```

### Get the Daily Weight Data

### Plot the Daily Weight Log

```python
# Plot creation
s = sns.set_context("paper")

plt.rc('axes', labelsize=9)    # fontsize of the x and y labels
plt.rc('axes', titlesize=9)  # fontsize of the figure title
plt.rc('xtick', labelsize=7)    # fontsize of the tick labels
plt.rc('ytick', labelsize=7)    # fontsize of the tick labels
plt.rc('legend', fontsize=10)    # legend fontsize

fig, ax = plt.subplots(figsize=(3.75,2), dpi=300, facecolor='w')

lin1 = ax.plot('Day', 'Weight', data=DailyWeightLogdf,color="blue", linewidth=.2 )

ax.set_xlabel('Day')
ax.set_ylabel('Weight',color="blue")

ax.margins(x = 0.2, y= .01)
# plt.subplots_adjust(left=0, bottom=0, right=2, top=1, wspace=0, hspace=0)

ax.tick_params(axis='y', colors='blue')
plt.xlim([0, 2501])
plt.ylim([180, 215])
plt.title('Weight by Day', fontsize=9)

plt.savefig('DailyWeightLog.png', dpi=300)

plt.show()
```
